## Histogram

Plot for quantitative value with feature values on the x-axis and counts on y-axis. The counts are not on single values but on value ranges.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

## Allows to render plots inlie
%matplotlib inline

pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
plt.hist(data = pokemon, x = 'speed');
```

```python
## Default is only 10 bins
plt.hist(data = pokemon, x = 'speed', bins = 20)
```

```python
## bins are off
bins = np.arange(0, pokemon["speed"].max() + 5, 5)
plt.hist(data = pokemon, x = 'speed', bins = bins);
```

```python
## With seaborn

sb.distplot(pokemon['speed']); ## Density plot
```

```python
sb.distplot(pokemon['speed'], kde = False); ## Density plot
```

```python
## Plotting side by side

# Resize the chart, and have two plots side-by-side
# Set a larger figure size for subplots
plt.figure(figsize = [20, 5]) 

# 1 row, 2 cols, subplot 1
plt.subplot(1, 2, 1) 
sb.distplot(pokemon['speed']); ## Density plot

# 1 row, 2 cols, subplot 2
plt.subplot(1, 2, 2)
sb.distplot(pokemon['speed'], kde = False); ## Density plot
```
